```python
#define a function to add the elements present in a tuple.tuple will be given as an argument
def new(tup1):
    new=0
    for i in tup1:
        new=new+i
    print(new)   
tup=(1,2,3,4)
new(tup)
```

```python
#using zip function make a list of tupples and then write a program to sort a list of tuples using lambda.
l1=[24,56,63,90]
l2=[85,12,89,12]
m=(list(zip(l1,l2)))
m.sort(key = lambda x: x[1])
print("According to second element")
print(m)
print("According to first element")
m.sort()
print(m)
```

```python
import pandas as pd
import numpy as np
data = pd.DataFrame({
     'Brand' : ['Maruti', 'Hyundai', 'Tata', 'Mahindra','Maruti', 'Hyundai', 'Renault', 'Tata', 'Maruti'],
     'Year' : [2012, 2014, 2011, 2015, 2012,2016, 2014, 2018, 2019],
     'Kms Driven' : [50000, 30000, 60000,25000, 10000, 46000,31000, 15000, 12000],
     'City' : ['Gurgaon', 'Delhi', 'Mumbai','Delhi', 'Mumbai', 'Delhi','Mumbai','Chennai', 'Ghaziabad'],
     'Mileage' : [28, 27, 25, 26, 28,29, 24, 21, 24]})
data
```

```python
#1.selecting cars with brand 'Maruti' and Mileage > 25
data.loc[(data.Mileage>25) & (data.Brand=='Maruti')]
```

```python
#2.selecting range of rows from 2 to 5
data.loc[2:5]
```

```python
#3.updating values of Mileage if Year < 2015
data.loc[(data.Year<2015),('Mileage')]= '23'
data
```

```python
#4.selecting 0th, 2th, 4th, and 7th index rows
data.loc[[0,2,4,7]]
```

```python
#5.selecting rows from 1 to 4 and columns from 2 to 4
data.iloc[1:5,2:4]
```
